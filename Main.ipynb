```python
skincare['tags'] = skincare['editors_note']+skincare['ingredients']+skincare['benefits']
```

```python
skincare
```

```python
skincare_new = skincare.drop(columns=['editors_note'])
skincare_new = skincare_new.rename(columns={'name': 'product_name'})
```

```python
skincare_new['tags'].str.count(' ')+1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(skincare_new['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarities = cosine_similarity(vector)
```

```python
similarities
```

```python
skincare_new[skincare_new['product_name']=='Torriden - Balanceful Cica Serum'].index[0]
```

```python
skincare_new.info()
```

```python
distances = sorted(list(enumerate(similarities[340])), reverse=True, key = lambda vector:vector[1])
for i in distances[1:6]:
    print(skincare_new.iloc[i[0]].product_name)
```

```python
def recommend(product, n):
    index = skincare_new[skincare_new['product_name']==product].index[0]
    distances = sorted(list(enumerate(similarities[index])), reverse=True, key = lambda vector:vector[1])
    if n + 1 < 357:
        n += 1
    else:
        n = 356
    for i in distances[1:n]:
        print(skincare_new.iloc[i[0]].product_name)
```

```python
recommend('Torriden - Balanceful Cica Serum', 5)
```

```python
vector2 = cv.fit_transform(skincare_new['ingredients'].values.astype('U')).toarray()
```

```python
vector2.shape
```

```python
ingredients_similarities = cosine_similarity(vector2)
```

```python
ingredients_similarities
```

```python
ingredients_distances = sorted(list(enumerate(ingredients_similarities[340])), reverse=True, key = lambda vector:vector[1])
for i in ingredients_distances[1:6]:
    print(skincare_new.iloc[i[0]].product_name)
```

```python
def find_similar_ingredients(product, n):
    index = skincare_new[skincare_new['product_name']==product].index[0]
    distances = sorted(list(enumerate(ingredients_similarities[index])), reverse=True, key = lambda vector:vector[1])
    if n + 1 < 357:
        n += 1
    else:
        n = 356
    for i in distances[1:n]:
        print(skincare_new.iloc[i[0]].product_name)
```

```python
find_similar_ingredients('Torriden - Balanceful Cica Serum', 7)
```
